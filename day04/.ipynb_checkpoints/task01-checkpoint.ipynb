#### Произведем необходимые преобразования над данными, чтобы с ними было удобно работать

```python
df.rename(index=str, columns={"charcter_1": "character_1"}, inplace=True)
df = df[~df['sr_start'].isin(['-'])]
df = df[~df['my_team_sr'].isin(['-'])]
df = df[~df['enemy_team_sr'].isin(['-'])]
df = df[~df['sr_finish'].isin(['-'])]
df = df[~df['my_team_2'].isin(['-'])]
df = df[~df['my_team_3'].isin(['-'])]
df = df[~df['my_team_4'].isin(['-'])]
df = df[~df['my_team_5'].isin(['-'])]
df = df[~df['my_team_6'].isin(['-'])]
df = df[~df['sr_delta'].isin(['-', '2500%'])]
df = df[~df['result'].isin(['-', 'DC'])]

df.loc[df['result'] == 'Loss', 'result'] = 'L'
df.loc[df['result'] == 'l', 'result'] = 'L'
df.loc[df['result'] == 'Draw', 'result'] = 'D'
df.loc[df['result'] == 'Win', 'result'] = 'W'

columns = ['win', 'lose', 'draw']
for col in columns:
    df[col] = 0
    
df.loc[df['result'] == 'W', 'win'] = 1
df.loc[df['result'] == 'L', 'lose'] = 1
df.loc[df['result'] == 'D', 'draw'] = 1
```

#### Посчитаем количество персонажей, чтобы удобней было выявить, какие данные надо подправить, чтобы не было похожих повторов (Lucio/Luico)

```python
df.character_1.value_counts()
```

```python
name_pairs = [ [ [ 'Luico', 'LUcio' ], 'Lucio' ], [ [ 'Mery', 'mercy' ], 'Mercy' ], [ [ 'Rienhart' ], 'Reinhardt' ], 
               [ [ 'Solider76' ], 'Soldier76' ], [ [ 'zen ', 'Zen ', 'zen' ], 'Zen' ], [ [ 'torbjorn' ], 'Torbjorn' ],
               [ [ 'ANa' ], 'Ana' ], [ [ 'zarya' ], 'Zarya' ] ]
```

```python
df = df[~df['character_1'].isin(['Illios', '-'])]
```

```python
for name_pair in name_pairs:
    for name in name_pair[0]:
        df.loc[df['character_1'] == name, 'character_1'] = name_pair[1]
```

#### Возьмем тестовый DF

```python
season_ratings_df = df[['season', 'sr_start', 'sr_finish', 'sr_delta', 'win', 'lose', 'draw', 'character_1']]
```

```python
season_ratings_df = season_ratings_df.dropna()
```

```python
season_ratings_df.shape
```

```python
season_ratings_df.sr_start = season_ratings_df.sr_start.astype('int64')
season_ratings_df.sr_finish = season_ratings_df.sr_finish.astype('int64')
season_ratings_df.sr_delta = season_ratings_df.sr_delta.astype('int64')
```

```python
delta_ratings = season_ratings_df[['season', 'sr_delta']].groupby('season').sum()
```

####  Попробуем взглянуть, как изменялся сезонный рейтинг игрока по сезонам
**sr_delta** - изменение рейтинга игрока в конце игры<br/>
**season** - игровой сезон

```python
delta_ratings.plot(kind='bar')
```

Видно, что в шестом сезоне игрок поднял больше всего рейтинга, то есть играл лучше, а в 3, 4 и 7 сезонах он в итоге закончил с меньшим рейтингом, чем начинал.

#### Теперь посмотрим на соотношения побед, поражений и ничейных результатов

```python
win_lose = season_ratings_df[['season', 'win', 'lose', 'draw']].groupby('season').sum()
```

```python
win_lose.plot(kind='bar', rot=45);
```

Положительная динамика побед к поражениям наблюдается только в 4 и 6 сезоне (что объясняет увеличение, а не уменьшение рейтинга в этих сезонах). Занятное наблюдение: в 4 сезоне игрок победил всего-лишь на несколько раз больше, чем проиграл, а рейтинг вырос на 200 с лишним пунктов. Видимо, за победу ему давали прилично больше рейтинга, чем отнимали за поражение.

#### Для наглядности взглянем на линейный график

```python
def plotly_line_plot(df, title = ''):
    data = []
    
    for column in df.columns:
        trace = go.Scatter(
            x = df.index,
            y = df[column],
            mode = 'lines',
            name = column
        )
        data.append(trace)
    
    layout = dict(title = title)
    fig = dict(data = data, layout = layout)
    iplot(fig, show_link=False)
```

```python
plotly_line_plot(
    win_lose,
    title = 'Player statistics'
)
```

#### Ну и напоследок рассмотрим его команду.

```python
init_notebook_mode(connected=True)
```

```python
teammates = df[ ['my_team_2', 'my_team_3', 'my_team_4', 'my_team_5', 'my_team_6'] ]
```

```python
teammates = teammates.melt()
```

```python
pd.crosstab(index=teammates['value'], columns=teammates['variable']).head(5)
```

#### Возьмем 5 наиболее частых его тиммейтов

```python
top_teammates = teammates.apply(pd.value_counts)['value'].nlargest(10).head(5)
```

```python
top_teammates = list(top_teammates.index.values)
```

```python
top_teammates
```

```python
for name in top_teammates:
    df[name] = 0
    positions = [ 'my_team_2', 'my_team_3', 'my_team_4', 'my_team_5', 'my_team_6' ]
    for position in positions:
        df.loc[df[ position ] == name, name] = 1
```

```python
df_maps = df[ [ 'sr_delta', 'map', 'Poke', 'Darkot', 'Yrova', 'HBHack', 'Looty'] ] 
```

```python
df_maps_teammates = df_maps.dropna()
```

```python
for name in top_teammates:  
    mask = df_maps_teammates[name] > 0
    column_name = name
    df_maps_teammates.loc[mask, column_name] = df_maps_teammates['sr_delta']
```

#### Взглянем, сколько игрок получает или теряет рейтинга с конкретным членом команды на конкретной карте

```python
df_maps_teammates = df_maps_teammates[ [ 'map', 'Poke', 'Darkot', 'Yrova', 'HBHack', 'Looty' ] ]
```

```python
df_maps_teammates.head(5)
```

```python
df_maps_teammates.info()
```

```python
for name in top_teammates:
    df_maps_teammates[name] = df_maps_teammates[name].astype('int64')
```

#### Заменим кое-где названия карт, чтобы совпадали

```python
map_pairs = [ [ [ "king's row", "king's Row" ], "King's Row" ], 
              [ [ "ilios", "Illios" ], "Ilios" ], 
              [ [ 'Gilbraltar' ], 'Gibraltar' ], 
              [ [ 'Eichenwade', 'EichenWalde' ], 'Eichenwalde' ], 
              [ [ 'Volskaya', 'Volkaya Industries' ], 'Volskaya Industries' ], 
              [ [ 'Lijang Tower', 'lijang Tower', 'Lijang', 'LIjang Tower', 'lijiang Tower' ], 'Lijiang Tower' ], 
              [ [ 'Hoirzon' ], 'Horizon' ], 
              [ [ 'JunkerTown', 'junkerTown' ], 'Junkertown' ] ]
```

```python
for map_pair in map_pairs:
    for map in map_pair[0]:
        df_maps_teammates.loc[df_maps_teammates['map'] == map, 'map'] = map_pair[1]
```

```python
teammates_maps_results_sum = df_maps_teammates.groupby('map').sum()
```

#### Построим хитмэп значений изменения рейтинга для каждого тиммейта на конкретных картах

```python
sns.heatmap(teammates_maps_results_sum, annot=True, fmt=".0f", 
            linewidths=0.5);
```
